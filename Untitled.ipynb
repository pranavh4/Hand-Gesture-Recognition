```python
img_labels = img_label.get_images_labels_list("./data/images/test")
img_labels = [i['image_label'] for i in img_labels]
labels = sorted(list(set(img_labels)))
for i in range(len(img_labels)):
    one_hot = [0 for x in range(len(labels))]
    one_hot[labels.index(img_labels[i])] = 1
    img_labels[i] = one_hot
img_labels = np.array(img_labels)
# img_labels = [labels.index(x) for x in img_labels]
img_labels
```

```python
IMG_SHAPE = (240,320,3)
```

```python
model = VGG16(input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
```

```python
model.save("vgg16.h5")
```

```python
new_model = Sequential()
model.trainable = False
new_model.add(model)
new_model.add(Flatten())
new_model.add(Dense(18, activation='softmax'))
```

```python
new_model.compile(optimizer='adam',loss='categorical_crossentropy')
```
